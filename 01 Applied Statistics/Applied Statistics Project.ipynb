```python
#CHECKING THE PRESENCE OF MISSING VALUES
for attr in data.columns:
    print(data[attr].isnull().sum())
    
#ANSWER: No missing values found in any data column.
```

```python
#5-POINT SUMMARY OF NUMERICAL ATTRIBUTES
#Attribute Age
print(data['age'].describe()[['min','25%','50%','75%','max']])
#Attribute BMI
print(data['bmi'].describe()[['min','25%','50%','75%','max']])
#Attribute Children
print(data['children'].describe()[['min','25%','50%','75%','max']])
#Attribute Charges
print(data['charges'].describe()[['min','25%','50%','75%','max']])
```
